```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('0Szew3X123.jpg',-1)
```

```python
plt.imshow(img)
```

```python
height, width = img.shape[:2]
print('height =', height)
print('width =', width)
resized = cv2.resize(img, (55, 100))
plt.imshow(resized)
```

```python
center=tuple(np.array(resized.shape[0:2])/2)
rot_mat = cv2.getRotationMatrix2D(center,90,1.0)
rotated = cv2.warpAffine(resized, rot_mat, resized.shape[0:2],flags=cv2.INTER_LINEAR)

plt.imshow(rotated)
```
